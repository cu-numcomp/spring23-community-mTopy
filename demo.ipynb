# Pycaret

Pycaret is a low code machine learning and data science library for python. Its goal is to make model creation faster and easier than before allowing for increased productivity. It draws from many other open source projects to allow for increased functionality and customization.



```python
# use of a venv is recommended as flask and pycaret use different versions of jinja2
import pycaret
from pycaret.datasets import get_data
from pycaret.regression import *
```

```python

#we are going to use some of the provided data to make this easier
data = get_data('insurance')
experiment = RegressionExperiment()
#we create a new experiment for regression that targets the charges on the data we imported
experiment.setup(data,target='charges',session_id=888)
```

```python
#we can train different models on the created experiment
best_model = experiment.compare_models()
```

```python
#we can now graph our model
plot_model(best_model, plot = 'error')
```

```python
plot_model(best_model, plot = 'learning')
```

```python
plot_model(best_model, plot = 'residuals')
```
